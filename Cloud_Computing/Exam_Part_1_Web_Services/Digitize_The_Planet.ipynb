# Task 3 - Digitize the Planet

Get information about the protected area with UUID '9a482446-cab1-4338-9df8-7bb627b339e8':

```python
uuid = '9a482446-cab1-4338-9df8-7bb627b339e8'
area_url = f'https://content.digitizetheplanet.org/api/protectedarea/{uuid}'
response = requests.get(area_url)

if response.status_code == 200:
    prot_area = response.json()
    print('The API call was successful!')
    display(prot_area)
else: 
    print(f'The API call was not  successful! Error Code: {response.status_code}')
```

## 3.1

Print name of the protected area with UUID '9a482446-cab1-4338-9df8-7bb627b339e8' with infos from the protected area response:

```python
print(f'The name of the protected area with UUID "9a482446-cab1-4338-9df8-7bb627b339e8" is\n' +
      f'"{prot_area.get('name')}".')
```

**Answer:** The name of the protected area with UUID "9a482446-cab1-4338-9df8-7bb627b339e8" is "Schleswig-Holsteinisches Wattenmeer".

## 3.2

Print name of the geometry source organization with infos from the protected area response:

```python
print(f'The name of the source organization for the geometry information is ' +
      f'"{prot_area.get('geometry_source_organization')}".')
```

**Answer:** The name of the source organization for the geometry information is "Bundesamt für Naturschutz".

## 3.3

Print category of protected area with infos from the protected area response:

```python
print(f'The category of the protected area is "{prot_area.get('category').get('protectedareacategory')}".')
```

**Answer:** The category of the protected area is "National Park".

## 3.4

Get information about the protected area rule with ID 149:

```python
id = 149
url = f'https://content.digitizetheplanet.org/api/protectedarearule/{id}/'
response = requests.get(url)

if response.status_code == 200:
    prot_area_rule = response.json()
    print('The API call was successful!')
    display(prot_area_rule)
else: 
    print(f'The API call was not  successful! Error Code: {response.status_code}')
```

Print activity, activity place, activity permission, and name of protected area (park) with infos from the protected area rule response:

```python
print(f'The activity of the rule with the ID 149 is "{prot_area_rule.get('activity'). get('activity')}".')
print(f'The activity place of the rule with the ID 149 is "{prot_area_rule.get('activityplace'). get('place')}".')
print(f'The activity permission of the rule with the ID 149 is "{prot_area_rule.get('activitypermission'). get('permission')}".')
```

Print name of protected area (park) with infos from the protected area rule response:

```python
park_uuid = prot_area_rule.get('park')
url = f"https://content.digitizetheplanet.org/api/protectedarea/{park_uuid}"
response = requests.get(url)

if response.status_code == 200:
    prot_area_park = response.json()
    print('The API call was successful!')
    #display(prot_area_park) -> not necessary here, as it is the same output as in the very beginning of task 3
else: 
    print(f'The API call was not  successful! Error Code: {response.status_code}')

print(f'The name of protected area (park) is "{prot_area_park.get('name')}".')
```

**Answer:** For the area rule with ID 149 the activity of the rule is "Camping", the activity place of the rule is "Total area of the territory", the activity permission of the rule is "forbidden", and the name of protected area (park) where the rule applies to is "Ferbitzer Bruch".
